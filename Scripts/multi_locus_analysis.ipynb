```python
#Importing Libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Analysis from data generated using julia notebook

```python
range_c = np.linspace(1, 20, 20, dtype = 'int')
range_a = np.round(np.linspace(1, 3, 20), 2)
shapes = ["linear", "exp", "expinv", "s"]
filepath = "C:/project/Masters_2023_DrHannaK/DataCSVs/"

f_dom_exp = pd.read_csv(filepath, "f_dom_exp.csv", header = 0, sep = ',').set_index(range_a).iloc[::-1,::-1].round(2)
m_dom_exp = pd.read_csv(filepath, "m_dom_exp.csv", header = 0, sep = ',').set_index(range_a).iloc[::-1,::-1].round(2)
t_dom_exp = pd.read_csv(filepath, "t_dom_exp.csv", header = 0, sep = ',').set_index(range_a).iloc[::-1,::-1].round(2)
f_rec_exp = pd.read_csv(filepath, "f_rec_exp.csv", header = 0, sep = ',').set_index(range_a).iloc[::-1,::-1].round(2)
m_rec_exp = pd.read_csv(filepath, "m_rec_exp.csv", header = 0, sep = ',').set_index(range_a).iloc[::-1,::-1].round(2)
t_rec_exp = pd.read_csv(filepath, "t_rec_exp.csv", header = 0, sep = ',').set_index(range_a).iloc[::-1,::-1].round(2)

f_dom_s = pd.read_csv(filepath, "f_dom_s.csv", header = 0, sep = ',').set_index(range_a).iloc[::-1,::-1].round(2)
m_dom_s = pd.read_csv(filepath, "m_dom_s.csv", header = 0, sep = ',').set_index(range_a).iloc[::-1,::-1].round(2)
t_dom_s = pd.read_csv(filepath, "t_dom_s.csv", header = 0, sep = ',').set_index(range_a).iloc[::-1,::-1].round(2)
f_rec_s = pd.read_csv(filepath, "f_rec_s.csv", header = 0, sep = ',').set_index(range_a).iloc[::-1,::-1].round(2)
m_rec_s = pd.read_csv(filepath, "m_rec_s.csv", header = 0, sep = ',').set_index(range_a).iloc[::-1,::-1].round(2)
t_rec_s = pd.read_csv(filepath, "t_rec_s.csv", header = 0, sep = ',').set_index(range_a).iloc[::-1,::-1].round(2)

f_dom_expinv = pd.read_csv(filepath, "f_dom_expinv.csv", header = 0, sep = ',').set_index(range_a).iloc[::-1,::-1].round(2)
m_dom_expinv = pd.read_csv(filepath, "m_dom_expinv.csv", header = 0, sep = ',').set_index(range_a).iloc[::-1,::-1].round(2)
t_dom_expinv = pd.read_csv(filepath, "t_dom_expinv.csv", header = 0, sep = ',').set_index(range_a).iloc[::-1,::-1].round(2)
f_rec_expinv = pd.read_csv(filepath, "f_rec_expinv.csv", header = 0, sep = ',').set_index(range_a).iloc[::-1,::-1].round(2)
m_rec_expinv = pd.read_csv(filepath, "m_rec_expinv.csv", header = 0, sep = ',').set_index(range_a).iloc[::-1,::-1].round(2)
t_rec_expinv = pd.read_csv(filepath, "t_rec_expinv.csv", header = 0, sep = ',').set_index(range_a).iloc[::-1,::-1].round(2)

f_dom_linear = pd.read_csv(filepath, "f_dom_linear.csv", header = 0, sep = ',').set_index(range_a).iloc[::-1,::-1].round(2)
m_dom_linear = pd.read_csv(filepath, "m_dom_linear.csv", header = 0, sep = ',').set_index(range_a).iloc[::-1,::-1].round(2)
t_dom_linear = pd.read_csv(filepath, "t_dom_linear.csv", header = 0, sep = ',').set_index(range_a).iloc[::-1,::-1].round(2)
f_rec_linear = pd.read_csv(filepath, "f_rec_linear.csv", header = 0, sep = ',').set_index(range_a).iloc[::-1,::-1].round(2)
m_rec_linear = pd.read_csv(filepath, "m_rec_linear.csv", header = 0, sep = ',').set_index(range_a).iloc[::-1,::-1].round(2)
t_rec_linear = pd.read_csv(filepath, "t_rec_linear.csv", header = 0, sep = ',').set_index(range_a).iloc[::-1,::-1].round(2)
```

# Recessive Defender Pop

```python
fig, ax = plt.subplots(nrows = 3, ncols = 4, figsize = (37.5,22.5), sharey = True, sharex = True, layout = 'tight')

sns.heatmap(f_rec_s, annot=f_rec_s, cmap = sns.color_palette('Reds', as_cmap=True), vmin = 0.0, vmax = 1.0, ax = ax[0,0])
sns.heatmap(f_rec_exp, annot=f_rec_exp, cmap = sns.color_palette('Reds', as_cmap=True), vmin = 0.0, vmax = 1.0, ax = ax[0,1])
sns.heatmap(f_rec_expinv, annot=f_rec_expinv, cmap = sns.color_palette('Reds', as_cmap=True), vmin = 0.0, vmax = 1.0, ax = ax[0,2])
sns.heatmap(f_rec_linear, annot=f_rec_linear, cmap = sns.color_palette('Reds', as_cmap=True), vmin = 0.0, vmax = 1.0, ax = ax[0,3])

sns.heatmap(m_rec_s, annot=m_rec_s, cmap = sns.color_palette("Blues", as_cmap=True), vmin = 0.0, vmax = 1.0, ax = ax[1,0])
sns.heatmap(m_rec_exp, annot=m_rec_exp, cmap = sns.color_palette("Blues", as_cmap=True), vmin = 0.0, vmax = 1.0, ax = ax[1,1])
sns.heatmap(m_rec_expinv, annot=m_rec_expinv, cmap = sns.color_palette("Blues", as_cmap=True), vmin = 0.0, vmax = 1.0, ax = ax[1,2])
sns.heatmap(m_rec_linear, annot=m_rec_linear, cmap = sns.color_palette("Blues", as_cmap=True), vmin = 0.0, vmax = 1.0, ax = ax[1,3])

sns.heatmap(f_rec_s.subtract(m_rec_s), annot=t_rec_s, cmap = sns.color_palette("RdBu_r", as_cmap=True), vmin = -1, vmax = 1, ax = ax[2,0])
sns.heatmap(f_rec_exp.subtract(m_rec_exp), annot=t_rec_exp, cmap = sns.color_palette("RdBu_r", as_cmap=True), vmin = -1, vmax = 1, ax = ax[2,1])
sns.heatmap(f_rec_expinv.subtract(m_rec_expinv), annot=t_rec_expinv, cmap = sns.color_palette("RdBu_r", as_cmap=True), vmin = -1, vmax = 1, ax = ax[2,2])
sns.heatmap(f_rec_linear.subtract(m_rec_linear), annot=t_rec_linear, cmap = sns.color_palette("RdBu_r", as_cmap=True), vmin = -1, vmax = 1, ax = ax[2,3])

ax[0,0].set_title('Sigmoid')
ax[0,1].set_title('Exponential')
ax[0,2].set_title('Inverse Exponential')
ax[0,3].set_title('Linear')

ax[2,0].set_xlabel('Value for c')
ax[2,1].set_xlabel('Value for c')
ax[2,2].set_xlabel('Value for c')
ax[2,3].set_xlabel('Value for c')

ax[0,0].set_ylabel('Value for a')
ax[1,0].set_ylabel('Value for a')
ax[2,0].set_ylabel('Value for a')

fig.suptitle('Proportion of defenders in both sexes: Recessive')
```
